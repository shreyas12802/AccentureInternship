```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
content = pd.read_csv("Content.csv")
reaction = pd.read_csv("Reactions.csv")
reaction_type =pd.read_csv("ReactionTypes.csv")
```

```python
new_reaction=reaction.merge(content,on="Content ID")
```

```python
new_reaction=new_reaction.merge(reaction_type,on="Type")
```

```python
new_reaction
```

```python
new_reaction['Category'] = new_reaction['Category'].str.lower()
category_sum=new_reaction.groupby("Category")["Score"].sum().reset_index(name="ScoreTotal")
category_sum=category_sum.sort_values(by="ScoreTotal",ascending=False)
category_sum
```

```python
new_reaction.to_excel("cleaned.xlsx",sheet_name='Sheet1', index=False)
```

```python
category_sum.to_excel("cleaned.xlsx",sheet_name='Category totals', index=False)
```

```python
new_reaction
```

```python
new_reaction["Datetime"].dtype
```

```python
new_reaction["Datetime"] = pd.to_datetime(new_reaction["Datetime"],format='mixed', dayfirst=True)
new_reaction["Month"] = new_reaction["Datetime"].dt.month_name()
```

```python
new_reaction.head()
```

```python
month_scores = new_reaction.groupby("Month")["Score"].count().reset_index(name="Reactions")
#month_scores = month_scores.sort_values(by="")
month_scores
```

```python
content_scores = new_reaction.groupby("Content Type")["Score"].count().reset_index(name="Reactions")
content_scores
```

```python
with pd.ExcelWriter("cleaned.xlsx") as writer:
   
    # use to_excel function and specify the sheet_name and index 
    # to store the dataframe in specified sheet
    new_reaction.to_excel(writer, sheet_name="Reactions", index=False)
    category_sum.to_excel(writer, sheet_name="Category totals", index=False)
    month_scores.to_excel(writer, sheet_name="Month Reactions", index=False)
    content_scores.to_excel(writer, sheet_name="Content Type Reactions", index=False)
```
